```python
img = cv2.imread(str(path_color))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
black_scan = cv2.imread(str(path_black))
threshould_value = 179
threshould = cv2.threshold(gray, threshould_value, 255,cv2.THRESH_BINARY_INV)[1]
bin_img_threshould = 1 - (threshould / 255.0)
gaussian = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 21, 4)
bin_img_gaussian = 1 - (gaussian / 255.0)
mean_c = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 21, 4)
bin_img_mean_c = 1 - (mean_c / 255.0)
```

```python
def compute_dense(dir_img, model):
    root = Path(r"C:\Users\Lenovo\Desktop\Sprayer Test _BackUp")
    ver = Path(model)
    color = Path("01_Water_sensitive_paper_color")
    black = Path("02_Water_sensitive_paper_Black_White")
    dir_image_name = Path(dir_img)

    path_color = root / ver / color / dir_image_name
    path_black = root / ver / black / dir_image_name

    list_color = os.listdir(path_color)
    list_black = os.listdir(path_black)

    dataset = []

    for color_img, black_img in zip(list_color, list_black):
        print(str(path_color / color_img))

        img = cv2.imread(str(path_color / color_img))
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        black_scan = cv2.imread(str(path_black / black_img))
        threshould_value = 179
        threshould = cv2.threshold(gray, threshould_value, 255,cv2.THRESH_BINARY_INV)[1]
        bin_img_threshould = 1 - (threshould / 255.0)
        gaussian = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 21, 4)
        bin_img_gaussian = 1 - (gaussian / 255.0)
        mean_c = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 21, 4)
        bin_img_mean_c = 1 - (mean_c / 255.0)

        w, h, _ = img.shape
        black_scan_dense = round(sum(black_scan.flatten() < 255)/ (w*h*3) * 100, 2)
        threshould_dense = round(sum(bin_img_threshould.flatten() == 0)/ (w*h) * 100, 2)
        gaussian_dense = round(sum(bin_img_gaussian.flatten() == 0)/ (w*h) * 100, 2)
        mean_c_dense = round(sum(bin_img_mean_c.flatten() == 0)/ (w*h) * 100, 2)

        dataset.append([color_img] + extract_name(color_img) + [black_scan_dense, threshould_dense, gaussian_dense, mean_c_dense])

    return dataset
```

```python
ls_dir_img_04 = os.listdir(r"C:\Users\Lenovo\Desktop\Sprayer Test _BackUp\04_AiANG_16L\01_Water_sensitive_paper_color")
ls_dir_img_05 = os.listdir(r"C:\Users\Lenovo\Desktop\Sprayer Test _BackUp\05_AiANG_30L\01_Water_sensitive_paper_color")
ls_dir_img = [ls_dir_img_04, ls_dir_img_05]
ls_model = os.listdir(r"C:\Users\Lenovo\Desktop\Sprayer Test _BackUp")
ls_dataset = {}
columns = ["File name", "Model (L)", "Rate (L/rai)", "Speed (m/s)", "Altitude (m)", "Distance (cm)", 
            "black_scan_dense (%)", "threshould_dense (%)", "gaussian_dense (%)", "mean_c_dense (%)"]
for n, model in enumerate(ls_model):
    set_dataset = {}
    dir_img = ls_dir_img[n]
    for dir in dir_img:
        dataset =  compute_dense(dir, model)
        dataset = np.array(dataset)
        set_dataset[dir] = pd.DataFrame(dataset, columns=columns)
    ls_dataset[model] = set_dataset
```
